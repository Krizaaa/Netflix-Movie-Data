```python
# Importing libraries 
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")

# Keeping only rows where the content type is 'Movie'
movies = (netflix_df["type"] == "Movie")

# Keeping movies released between 1990 and 1999 "inclusive"
release_1990s = (netflix_df["release_year"] >= 1990) & (netflix_df["release_year"] <= 1999)

# Selecting the movies that belong to the genre "Action"
action_movie = (netflix_df["genre"] == "Action")

# Combine all filters to get action movies from the 1990s
movies_release_90s = netflix_df[movies & release_1990s & action_movie]

# Initialize counter for short movies
short_movie_count = 0

# Loop through the filtered DataFrame (NOT the boolean filter)
for index, row in movies_release_90s.iterrows():
    
    # Check if the movie duration is less than 90 minutes
    if row["duration"] < 90:
        short_movie_count += 1  # Increment counter

# Print the final count
print(short_movie_count)
```
